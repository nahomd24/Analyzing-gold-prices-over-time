# Analyze Gold Prices Over Time

```python
# Nahom Diress
import pandas as pd
import seaborn as sns
```

```python
# 1) Read the data from the CSV file into a DataFrame called gold.
gold = pd.read_csv("gold.csv")
```

```python
# 2) Run the info() method, and note the data types of the two columns.
gold.info()
```

```python
# 3) Convert the Date column to the DateTime data type. 
# display the first five rows of data, and note that the dates are for the first of each month starting in 1950.
gold['Date'] = pd.to_datetime(gold['Date'])
```

```python
#4) Plot the data with a Pandas line plot.
gold.plot.line(y='Price')
```

```python
# 5) Index the data on the Date column. Then, generate time periods for the first day of each year from 1950 to 2020.
gold.set_index('Date', inplace=True)
gold.head(3)
```

```python
pd.date_range('01/1950', '01/2020', freq= 'YS')
```

```python
#6) Reindex the data using the new time periods, and display the first five rows again to see that there is just one row for each year.
gold_data_ys = gold.reindex(pd.date_range(start='1950-01-01', end='2020-01-01', freq='YS'))
gold_data_ys.head()
```

```python
# 7 Plot the reindexed data with a Pandas line plot.
gold_data_ys.plot.line(y='Price')
```

```python
# 8 Resample the data to a quarterly frequency and plot the data again.
gold_data_ys.resample(rule='QS').sum()
gold_data_ys.plot.line(y='Price')
```

```python
#9 Resample the data to a yearly frequency and plot the data one more time.
gold_data_ys.resample(rule='YS')
gold_data_ys.plot.line(y='Price')
```

```python
# 10) Plot the rolling mean for the data using a Pandas line plot.
df = gold[['Price']]. \
query('Date <= "01/31/2021"').rolling(window=7, min_periods=1).mean() 
df.head(8) 
```

```python
df.plot()
```
